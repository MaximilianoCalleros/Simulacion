```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Paso 1: Generar 100 números aleatorios
n = 100
random_numbers = np.random.uniform(0, 1, n)

# Paso 2: Realizar la prueba de Kolmogorov-Smirnov
ks_statistic, p_value = stats.kstest(random_numbers, 'uniform')

# Paso 3: Imprimir los resultados
print(f"Estadístico KS: {ks_statistic}")
print(f"Valor p: {p_value}")

# Paso 4: Interpretar los resultados
alpha = 0.05  # Nivel de significancia del 5%
if p_value > alpha:
    print("No se rechaza la hipótesis nula (distribución uniforme).")
else:
    print("Se rechaza la hipótesis nula (no es una distribución uniforme).")

# Paso 5: Graficar la distribución de los números generados
plt.hist(random_numbers, bins=10, edgecolor='black', alpha=0.7)
plt.title('Histograma de los números aleatorios generados')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```
