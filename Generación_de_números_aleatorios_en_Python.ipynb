#Método de centros al cuadrado

```python
def centros_al_cuadrado(semilla, cantidad):
    # Inicializamos una lista para almacenar los números generados
    numeros = []
    n = semilla

    for _ in range(cantidad):
        # Elevar al cuadrado el número
        n2 = str(n**2).zfill(8)  # Zfill para asegurar que tenga al menos 8 dígitos
        # Tomar los 4 dígitos centrales
        n = int(n2[len(n2)//2-2:len(n2)//2+2])
        # Guardar el número generado en la lista
        numeros.append(n)

    return numeros

# Generar 100 números aleatorios con una semilla
semilla_inicial = 23498
cantidad_numeros = 100
numeros_aleatorios = centros_al_cuadrado(semilla_inicial, cantidad_numeros)

# Imprimir los números generados
for i, num in enumerate(numeros_aleatorios):
    print(f"Número {i+1}: {num}")
```

#Método de medios cuadrados

```python
def medios_cuadrados(semilla, cantidad):
    # Lista para almacenar los números generados
    numeros = []
    n = semilla
    digitos = len(str(semilla))  # Número de dígitos de la semilla

    for _ in range(cantidad):
        # Elevar al cuadrado el número
        n2 = str(n ** 2).zfill(2 * digitos)  # Cuadrado de la semilla con relleno de ceros
        # Tomar los dígitos centrales
        mitad = len(n2) // 2
        n = int(n2[mitad - digitos // 2:mitad + digitos // 2])
        # Guardar el número en la lista
        numeros.append(n)

    return numeros

# Generar 100 números aleatorios usando una semilla
semilla_inicial = 5735
cantidad_numeros = 100
numeros_aleatorios = medios_cuadrados(semilla_inicial, cantidad_numeros)

# Imprimir los números generados
for i, num in enumerate(numeros_aleatorios):
    print(f"Número {i+1}: {num}")
```
