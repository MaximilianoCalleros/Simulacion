```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial (tasa de llegadas)
lambda_llegadas = 1 / 15  # Tiempo promedio entre llegadas = 15 minutos

# Función para generar un tiempo entre llegadas
def generar_tiempo_entre_llegadas(lambda_llegadas):
    u = np.random.uniform(0, 1)  # Generar número aleatorio uniforme
    tiempo_entre_llegadas = -1 / lambda_llegadas * np.log(1 - u)  # Aplicar transformada inversa
    return tiempo_entre_llegadas

# Generar tiempos de llegada durante un día de operación (12 horas = 720 minutos)
num_llegadas = int(720 / 15)  # Número esperado de llegadas
tiempos_entre_llegadas = [generar_tiempo_entre_llegadas(lambda_llegadas) for _ in range(num_llegadas)]

# Mostrar los tiempos de llegada generados
print("Tiempos entre llegadas de vehículos eléctricos (en minutos):")
print(tiempos_entre_llegadas)

# Graficar los tiempos de llegada simulados
plt.hist(tiempos_entre_llegadas, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos entre Llegadas de Vehículos Eléctricos (Exponencial)')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```
